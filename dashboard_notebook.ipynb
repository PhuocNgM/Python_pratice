### Bước 1: Khởi động Streamlit Server
Chạy cell dưới đây để khởi động server Streamlit chạy ngầm. Giao diện dashboard sẽ được phục vụ tại `http://localhost:8501`.

### Nhận xét & Giải thích Kết quả Dashboard
Phần Streamlit Dashboard ở trên cung cấp 4 tab chức năng chính để phân tích:
- **Bản đồ Giao thông (Traffic Map)**: Hiển thị trực quan trạng thái của từng đoạn đường. Điểm màu đỏ và biểu tượng vòng tròn lớn (Congested) cho thấy khu vực có vận tốc thấp, đang xảy ra ùn tắc nghiêm trọng. Có thanh trượt thời gian (Time Layer) để tua nhanh diễn biến giao thông.
- **Bản đồ Nhiệt (Heatmap)**: Cho phép chuyển đổi giữa Lớp Mật độ (Density), Vận tốc (Velocity) và Mức độ Phục vụ (LOS). Ví dụ lớp Mật độ sử dụng thang màu từ Nhạt (thấp) đến Điểm sáng (rất cao) để làm nổi bật các "tụ điểm" đông đúc. Vùng càng rực sáng ở trung tâm thành phố thì áp lực giao thông càng lớn.
- **Luật Kết hợp (Association Rules)**: Đây là đồ thị biểu diễn mối liên hệ ùn tắc giữa các đoạn đường. Nút (node) có kích thước lớn thể hiện đây là các đoạn đường giao cắt trọng yếu. Đường nối (edge) càng sáng và càng lớn chứng tỏ **độ tin cậy (Confidence)** của việc đoạn đường này tắc kéo theo đoạn đường kia tắc là càng cao.
- **Báo cáo Tự động (Insights Report)**: Tự động tổng hợp và tóm tắt các điểm nóng giao thông, phân tích khung giờ cao điểm và đánh giá phân phối LOS tổng thể.
